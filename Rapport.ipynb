```python
import matplotlib.pyplot as plt
```

### Blocks

```python
# Nombre d'étapes 
SAT_et = [8,10,8,16,14]
HSP_et = [6,10,6,12,10]


# Temps d'exéc
SAT_tmp = [11,13,12,24,18]
HSP_tmp = [6,6,5,8,6]


fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Graphique d'étapes
ax1.plot(HSP_et, label='HSP', color='blue')
ax1.plot(SAT_et, label='SAT', color='red')
ax1.set_xlabel('difficulté du problème')
ax1.set_ylabel('Nombre d\'étapes')
plt.ylabel('Nombre d\'étapes')
ax1.set_xticks(range(5))
ax1.set_xticklabels(range(1, 6))
ax1.set_title('Blocks problem')
ax1.legend()

# Graphique de temps d'exécution
ax2.plot(HSP_tmp, label='HSP', color='blue')
ax2.plot(SAT_tmp, label='SAT', color='red')
ax2.set_xlabel('difficulté du problème')
plt.ylabel('Temps d\'exécution')
ax2.set_xticks(range(5))
ax2.set_xticklabels(range(1, 6))
ax2.set_title('Blocks problem')
ax2.legend()

# Affichage des graphes
plt.show()
```

On peut voir la comparaison à gauche entre les nombres d'étapes des plans solution pour HSP et SAT, à droite la comparaison des temps de résolution pour HSP et SAT en dixièmes de secondes.
