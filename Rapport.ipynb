## Exercice 2 : SAT encoding

### Quel est le lien entre la formule SAT à résoudre et son encodage SAT ?
L'encodage transforme la formule SAT pour qu'un solveur puisse la comprendre, tout en gardant le même problème à résoudre. La formule SAT est un ensemble de clauses, son encodage permet de remplacer chaque clause par une variable afin d’obtenir des formes normales conjonctive.

### Les formules 3SAT encodées seront-elles plus difficiles à résoudre que les instances originales ?
 L'encodage peut avoir un effet sur la difficulté de résolution, étant donné que la génération de variable est aléatoire, le problème peut donc être plus simple si l’encodage a un ratio clauses variable élevé ou faible ou plus difficile si le ratio clause/variable est moyen. En effet un problème avec peu de contrainte à satisfaire va logiquement être facile a résoudre, a l’opposé beaucoup de contrainte va limiter le nombre d’action possible ce qui permet de le résoudre aisément également. Alors qu'à part équivalente le problème aura beaucoup plus d’action possible et sera donc plus complexe à résoudre. On ne peut donc pas dire que les formules 3-SAT soient plus difficiles ou plus faciles, cela dépendra du problème.
